```python
df.drop(columns=['Not_Useful_Column'], inplace = True)
df.head(5)
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True )
df.head(5)
df.shape
```

```python
df['First_Name'] = df['First_Name'].str.strip()
df['Last_Name'] = df['Last_Name'].str.strip()
```

```python
df['Last_Name']
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/._')
```

```python
# df['Last_Name'] = df['Last_Name'].str.replace('/', ' ')
# df['Last_Name'] = df['Last_Name'].str.replace('...', ' ')
# df['Last_Name'] = df['Last_Name'].str.replace('_', ' ')
```

```python
# df['Phone_Number'] = df['Phone_Number'].str.replace('|',"")
# df['Phone_Number'] = df['Phone_Number'].str.replace('/',"")
# df['Phone_Number'] = df['Phone_Number'].str.replace('-',"")
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[|/-]', '', regex=True)
df = df[(df['Phone_Number']!="") & (df['Phone_Number']!="Na") & (df['Phone_Number'].notna())]
df['Phone_Number']
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df = df[df['Paying Customer'] != 'N/a']
df['Paying Customer']
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace("No","N",)
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace("Yes","Y",)
df = df[df['Do_Not_Contact'].notna() & (df['Do_Not_Contact'] != "")]
df['Do_Not_Contact']
```
