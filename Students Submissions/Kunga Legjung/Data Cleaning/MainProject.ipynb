```python
upper_limit = df["resolution_time_hours"].quantile(0.99)
df = df[df["resolution_time_hours"] <= upper_limit]
```

```python
df["status"] = (
    df["status"]
    .str.strip()
    .str.replace("_", " ")
    .str.title()
)
```

```python
df["ticket_age_days"] = (
    pd.Timestamp.now() - df["created_at"]
).dt.days
```

```python
df["resolution_bucket"] = pd.cut(
    df["resolution_time_hours"],
    bins=[0, 12, 24, 48, 72, 168, 1000],
    labels=["<12h", "12–24h", "1–2 days", "2–3 days", "3–7 days", "7+ days"]
)
```

```python
df["csat_percent"] = (df["csat_score"] / 5) * 100
```

```python
top_regions = df["region"].value_counts().nlargest(5).index
df["region_grouped"] = df["region"].where(
    df["region"].isin(top_regions),
    "Other"
)
```

```python
df["sla_breached"] = np.where(
    df["resolution_time_hours"] > 48,
    "Yes",
    "No"
)
```

```python
df.sort_values("created_at", inplace=True)
```

```python
df.rename(columns={
    "resolution_time_hours": "Resolution Time (Hours)",
    "csat_score": "CSAT Score",
    "customer_segment": "Customer Segment",
    "product_area": "Product Area"
}, inplace=True)
```

```python
df.to_csv(
    "cleaned_it_support_tickets.csv",
    index=False
)
```
